## **Convering date time** ##

```python
df=pd.read_csv("./data/employees.csv",parse_dates=['Start Date','Last Login Time'])   ### converting into date time object
df.info()
```

```python
df.head()
```

```python
### converting datatypes to reduce memory
df["Senior Management"]=df["Senior Management"].astype('bool')
df["Gender"]=df["Gender"].astype('category')
df.info()                      
```

```python
### HUGE MEMORY OPTIMIZTION FROM 62KB TO 49KB

```

## **FILTERING THE DATAFRAME** ##

```python
df.head()
```

```python
cond1=df["Gender"]=="Male"
cond2=df["Team"]=="Marketing"
len(df[cond1 & cond2])
```

```python
#Salary less then 90000
df[df["Salary"]<90000].head()
```

```python
#Finding the employees who are either from Senior Management or are working before 1 jan 1990
condition=(df["Senior Management"]==True) | (df["Start Date"]>"1990-01-01")
df[df["Start Date"]<"1990-01-01"].head()
```

```python
# Find a person named Robert from Client Services team
#or else retrieve the record of people who joined before 1 jun 2016

cond1=(df["First Name"]=="Robert") & (df["Team"]=="Client Services")
cond2=df["Start Date"]<"2016-06-01"
df[cond1 | cond2].head()
```

```python
### solving null

df[df['Team'].isnull()].head()
```

```python
df[df["Gender"].notnull()].head()
```

```python
### find the employees with salary bw 60000 and 70000
df[df["Salary"].between(60000,70000)].head()
```
