# Visualisation in Python - Matplotlib

## Bar Graph: Plotting sales across each product category

```python
# Creating an array with the sales amount
# Furniture: 4110451.90
# Technology: 4744557.50
# Office Supplies: 3787492.52
import numpy as np
products=np.array(['Furniture','Technology','Office Supplies'])
sales=np.array([4110451.90,4744557.50,3787492.52])
```

```python
import matplotlib.pyplot as plt
plt.bar(products,sales)
```

```python
#adding title x_comp and y_comp
plt.bar(products,sales,color='green',edgecolor='black')      ## adding color to the bar grapgh
plt.title('Sales across product categories',fontdict={'fontsize':20,'fontweight':5,'color':'Red'})
plt.xlabel('Products',fontdict={'fontsize':10,'fontweight':3,'color':'Green'})
plt.ylabel('Sales',fontdict={'fontsize':10,'fontweight':3,'color':'Green'})
plt.show()
```

```python
##Modifying ticks or marks on the y-axis
#plt.bar(products,sales,color='green',edgecolor='black')      ## adding color to the bar grapgh
plt.title('Sales across product categories',fontdict={'fontsize':20,'fontweight':5,'color':'Red'})
plt.xlabel('Products',fontdict={'fontsize':10,'fontweight':3,'color':'Green'})
plt.ylabel('Sales',fontdict={'fontsize':10,'fontweight':3,'color':'Green'})

tick_values=np.arange(0,7000000,1000000)
tick_labels=[str(i)+"L" for i in range(0,70,10)]
plt.yticks(tick_values,tick_labels)    ##plt.yticks(values,new_labels)

##modifying width of the bars
plt.bar(products,sales,color='green',edgecolor='black',width=0.4)
plt.show()
```

## Complete process for scatter plotting

```python
# creating a list of different categories and colors
product_categories=np.unique(product_category)
colors=['red','cyan','green']
# plotting the scatterplot with color coding the points belonging to different categories 
for color,category in zip(colors,product_categories):
    sales_category=sales[product_category==category]
    profit_category=profit[product_category==category]
    plt.scatter(profit_category,sales_category,c=color,label=category)

# Adding and formatting title
plt.title('Sales across Profits in various countries for different product categories',fontdict={'fontsize':20,'fontweight':5,'color':'Black'})

# Labeling Axes
plt.xlabel('Profit',fontdict={'fontsize':10,'fontweight':3,'color':'Green'})
plt.ylabel('Sales',fontdict={'fontsize':10,'fontweight':3,'color':'Green'})

# Adding legend for interpretation of points
plt.legend()  #to interpret the points from the category passed in scatter.

plt.show()
```

```python
fig,ax=plt.subplots(2,ncols=3,sharey=True)  ### ncols=3 means 3 different graphs and ax will be an array of size 3
                                          ### sharey=true means there will be 1 common y axis
africa, =ax[0,0].plot(years,sales_Africa)
africa.set_label('Africa')

asia, =ax[1,1].plot(years,sales_Asia_Pacific) ### type is a matplotlib 2-d object
asia.set_label('Asia')


europe, =ax[1,2].plot(years,sales_Europe)
europe.set_label('Europe')

#plt.legend()  ### this is wrong-->plt.legend([subplot_1, subplot_2, subplot_3],["label 1", "label 2", "label 3"])
plt.legend([africa, asia, europe],["Africa", "Asia","Europe"])
plt.show()
```
